### 1. МНК-оценки параметров $a$ и $b$

Необходимо минимизировать функционал ошибки:

$$
Q = \sum_{i = 1}^{200}{\epsilon_{i}^{2}} = \sum_{i = 1}^{200}{(Y_i - a - b \cdot t_i)^{2}} \rightarrow \min_{a, b}
$$

Для этого найдём дифференциал функционала ошибки по каждому параметру:

$$
\begin{equation}
\begin{cases}
\frac{\Delta Q}{\Delta a} = -2 \cdot \sum_{i = 1}^{200}{(Y_i - a - b \cdot t_i)} = 0\\\\
\frac{\Delta Q}{\Delta b} = -2 \cdot \sum_{i = 1}^{200}{t_i \cdot (Y_i - a - b \cdot t_i)} = 0
\end{cases}
\end{equation}
$$

Решив эту систему уравнений, получим оценённые параметры:

$$
\begin{equation}
\begin{cases}
\hat{a} = \overline{Y} - \hat{b} \cdot \overline{t}\\\\
\hat{b} = \frac{\sum_{i = 1}^{200}{(t_i - \overline{t})(Y_i - \overline{Y})}}{\sum_{i = 1}^{200}{(t_i - \overline{t})^2}}
\end{cases}
\end{equation}
$$
